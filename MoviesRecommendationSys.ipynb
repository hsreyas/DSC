```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import os
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection and processing

movies_data = pd.read_csv('movies.csv')
movies_data
```

```python
movies_data.shape
```

```python
#selecting relevent features

selected_features = ['genres','keywords','tagline','cast','director','original_language','overview']
selected_features
```

```python
#replacing the null values

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
#combining the columns

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+ movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['original_language']+' '+movies_data['overview']
```

```python
combined_features
```

```python
#coverting to the vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#similarity will give some similarity value by which we will be getting a similarity score
#it will compare eg avatar to all the other movies accordingly
```

```python
movie = input('enter your favourite movie name : ')
```

```python
#creating the list which contains all the movie titles present in data set

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#here we will find the close match for the movie privided by user

find_close_match = difflib.get_close_matches(movie, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the movie index with help of title

index_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_movie)
```

```python
#getting a list of similar movies

sim_score = list(enumerate(similarity[index_movie]))
print(sim_score)
```

```python
#now after getting the sim score for all the movies 
#its time to recommend more similar movies to the user 
```

```python
sorted_similar_movies = sorted(sim_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('You may also like : \n')

i = 1
for movies in sorted_similar_movies:
    index = movies[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if (i<16):
        print(i,'.',title_from_index)
        i+=1
```

```python
#just pasted the above code together so it will process alltogether

movie = input('enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie, list_of_all_titles)

close_match = find_close_match[0]

index_movie = movies_data[movies_data.title == close_match]['index'].values[0]

sim_score = list(enumerate(similarity[index_movie]))

sorted_similar_movies = sorted(sim_score, key = lambda x:x[1], reverse = True)

print('You may also like : \n')

i = 1

for movies in sorted_similar_movies:
    index = movies[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if (i<16):
        print(i,'.',title_from_index)
        i+=1
```
